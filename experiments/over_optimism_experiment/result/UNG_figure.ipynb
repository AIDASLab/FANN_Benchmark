```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# -----------------------------
# 1) 데이터셋 클래스 정의
# -----------------------------
GENERATOR_KEYS = {"sift_high", "sift_low", "gist_high", "gist_low"}
SYNTH_KEYS     = {"sift1m_ACORN", "sift1m_UNG", "sift1m_NHQ", "sift1m_RWalks"}
# 나머지는 모두 semi-real

# -----------------------------
# 2) 클래스별 스타일 (선/마커 통일)
# -----------------------------
STYLE_BY_CLASS = {
    "generator": {"linestyle": "-",  "marker": "D", "linewidth": 6, "markersize": 12, "edge": None, "markeredgewidth" : 0},  # 초록계열 (나중에 색만 바뀜)
    "synthetic": {"linestyle": "-", "marker": "D", "linewidth": 6, "markersize": 12, "edge": None, "markeredgewidth" : 0},  # 파랑계열
    "semi":      {"linestyle": "-", "marker": "D", "linewidth": 6, "markersize": 12, "edge": None, "markeredgewidth" : 0},  # 붉은계열
}

# -----------------------------
# 3) dataset_key → label 매핑 (건드리지 말라고 한 부분)
# -----------------------------
LABEL_MAP = {
    "sift_high":   "Generated Sift Low",
    "sift_low":    "Generated: Sift High",
    "gist_low":    "Generated Gist High",
    "gist_high":   "Generated: Gist Low",
    "sift1m_UNG":  "Synthetic: UNG",
    "sift1m_RWalks": "Synthetic: RWalks",
    "sift1m_NHQ":  "Synthetic: NHQ",
    "sift1m_ACORN": "Synthetic: ACORN",
    "arxiv": "Semi-Real: arxiv",
    "LAION1M": "Semi-Real: LAION1M",
    "tripclick": "Semi-Real: tripclick",
    "yfcc": "Semi-Real: YFCC",
}

# -----------------------------
# 4) 베이스 컬러 & 그라데이션 생성 함수
# -----------------------------
# BASE_COLOR_BY_CLASS = {
#     "synthetic":     "steelblue",      # 0~3
#     "semi":          "salmon",         # 4~7
#     "generator":     "mediumseagreen", # 8~11
# }


BASE_COLOR_BY_CLASS = {
    "synthetic":     "blue",      # 0~3
    "semi":          "red",         # 4~7
    "generator":     "green", # 8~11
}

# def make_grad_colors(base_color, n):
#     """
#     base_color RGB는 유지하고, alpha 값만 0.4~1.0 범위로 변화시키는 RGBA 리스트 반환.
#     """
#     base_rgb = mcolors.to_rgb(base_color)

#     # 0.4 ~ 1.0 사이에서 alpha 변화
#     alphas = np.linspace(0.4, 1.0, max(1, n))

#     colors = []
#     for a in alphas:
#         colors.append((*base_rgb, a))  # (r, g, b, alpha)
#     return colors
def make_grad_colors(base_color, n): 
    """ 
    base_color를 기준으로 n개의 밝기 다른 색을 만든다. 흰색과 섞어서 점점 밝아지는 톤의 리스트를 반환. 
    """ 
    base_rgb = np.array(mcolors.to_rgb(base_color)) 
    white = np.ones(3) # 0.4 ~ 1.0 사이에서 섞기 (0.4: 많이 하양, 1.0: 원색에 가까움) 
    alphas = np.linspace(0.2, 1.0, max(1, n)) 
    colors = [] 
    for a in alphas: 
        rgb = a * base_rgb + (1 - a) * white 
        colors.append(tuple(rgb)) 
    return colors
# -----------------------------
# 5) trade_off를 클래스별로 그룹핑
# -----------------------------
generator_keys = [k for k in trade_off.keys() if k in GENERATOR_KEYS]
synthetic_keys = [k for k in trade_off.keys() if k in SYNTH_KEYS]
semi_keys      = [k for k in trade_off.keys() if k not in GENERATOR_KEYS | SYNTH_KEYS]

GROUPS = {
    "generator": generator_keys,
    "synthetic": synthetic_keys,
    "semi":      semi_keys,
}

# 클래스별 색 리스트 미리 생성
CLASS_COLORS = {}
for cls, keys in GROUPS.items():
    base = BASE_COLOR_BY_CLASS[cls]
    CLASS_COLORS[cls] = make_grad_colors(base, len(keys))


# -----------------------------
# 6) 플롯
# -----------------------------
plt.figure(figsize=(12, 10))

for cls, keys in GROUPS.items():
    style = STYLE_BY_CLASS[cls]
    colors = CLASS_COLORS[cls]

    for local_idx, dataset_key in enumerate(keys):
        stats_dict = trade_off[dataset_key]
        label = LABEL_MAP.get(dataset_key, dataset_key)
        color = colors[local_idx]
        if label == "Synthetic: UNG":
            # color = "Blue"
            # style["linewidth"] = 7
            # style["markersize"] = 15
            style['edge'] = "black"
            style['markeredgewidth'] = 3
            z = 10
        else:
            style['edge'] = None
            style['markeredgewidth'] = 0
            z = 5
        # K_ns 정렬 (키가 "10", "20" 같은 문자열 숫자라고 가정)
        K_ns = sorted(stats_dict.keys(), key=lambda x: int(x))

        qps_values = [stats_dict[k]["qps"] for k in K_ns]
        recall_values = [stats_dict[k]["avg_recall"] for k in K_ns]

        plt.plot(
            qps_values,
            recall_values,
            marker=style["marker"],
            markersize=style["markersize"],
            markeredgecolor=style['edge'],
            markeredgewidth=style['markeredgewidth'],
            linestyle=style["linestyle"],
            linewidth=style["linewidth"],
            color=color,
            label=label,
            zorder=z
        )

plt.xscale("log")
plt.xlabel("QPS", fontsize=24, fontweight='bold' )
plt.ylabel("Recall", fontsize=24,fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("UNG", fontsize=32, fontweight='bold')
plt.legend(
    fontsize=16,
    ncol=3,
    bbox_to_anchor=(0.5, -0.15),
    loc="upper center",
)
plt.grid(True, which="both", linestyle=":", linewidth=0.7)
plt.tight_layout()
plt.savefig("UNG_figure.pdf")
plt.show()
```

```python
import numpy as np
import pandas as pd
from collections import Counter

# --- 설정 변수 ---
N_ROWS = 1000000            # 생성할 로우(아이템) 수
LABEL_POOL_SIZE = 12      # 전체 레이블 개수 (|L| = 12)
ZIPF_ALPHA = 1.0          # Zipf 분포의 지수 (Alpha). 

# --- 1. Zipfian 확률 계산 ---
labels = np.arange(1, LABEL_POOL_SIZE + 1)
probabilities_unnormalized = 1 / (labels ** ZIPF_ALPHA)
probabilities = probabilities_unnormalized / np.sum(probabilities_unnormalized)

print(f"**레이블별 선택 확률 (Zipf Alpha={ZIPF_ALPHA})**")
for l, p in zip(labels, probabilities):
    print(f"Label {l}: {p*100:.2f}%", end=' ' if l % 3 != 0 and l != 12 else '\n')
print("-" * 50)


# --- 2. 10,000개의 로우 데이터 생성 ---
generated_labels = []

for _ in range(N_ROWS):
    cardinality = np.random.randint(1, LABEL_POOL_SIZE + 1)
    
    assigned_labels = np.random.choice(
        labels, 
        size=cardinality, 
        replace=False, 
        p=probabilities
    )
    
    # ,로 구분된 문자열로 포맷팅
    label_string = ",".join(map(str, sorted(assigned_labels)))
    generated_labels.append(label_string)

# --- 3. 텍스트 파일 저장 (새로 추가된 부분) ---
FILENAME = "/home/ec2-user/hybrid_hardness/Benchmark/sift1m_UNG_modi/mid_format/base_label_UNG.txt"

try:
    with open(FILENAME, 'w') as f:
        # 각 레이블 세트 문자열 끝에 개행 문자('\n')를 붙여서 파일에 씁니다.
        for label_set in generated_labels:
            f.write(label_set + '\n')

    print(f"\n[INFO] 레이블 세트가 성공적으로 저장되었습니다: {FILENAME}")

except Exception as e:
    print(f"\n[ERROR] 파일 저장 중 오류가 발생했습니다: {e}")


# --- 4. 분포 검증 (선택적 출력) ---
all_labels_flat = []
for row_str in generated_labels:
    all_labels_flat.extend(map(int, row_str.split(',')))

label_counts = Counter(all_labels_flat)
total_occurrences = sum(label_counts.values())

print("\n[레이블 ID별 총 등장 비율 검증 결과]")
for label_id in sorted(label_counts.keys()):
    count = label_counts[label_id]
    ratio = (count / total_occurrences) * 100
    print(f"Label {label_id:2d}: {ratio:.4f}% ({count}회)")
```

```python
def load_groundtruth_bin(filename, nq):
    record_dtype = np.dtype([("idx", np.uint32), ("dist", np.float32)])
    file_size = os.path.getsize(filename)
    record_size = record_dtype.itemsize  # 8 bytes (4+4)
    total_recs = file_size // record_size

    if nq == 0 or total_recs % nq != 0:
        raise ValueError(f"GT file size mismatch: total_recs={total_recs}, nq={nq}")

    K_gt = total_recs // nq

    data = np.fromfile(filename, dtype=record_dtype, count=total_recs)
    # 인덱스만 추출, shape: (nq, K_gt)
    gt_indices = data["idx"].reshape(nq, K_gt)
    return gt_indices

gt_indices= load_groundtruth_bin("/home/ec2-user/hybrid_hardness/Benchmark/sift1m_UNG_modi/mid_format/gt.bin" ,10000)
gt_indices = gt_indices.tolist()
```
