```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# 1) 데이터셋 클래스 정의
# -----------------------------
GENERATOR_KEYS = {"sift_high", "sift_low", "gist_high", "gist_low"}
SYNTH_KEYS     = {"sift1m_ACORN", "sift1m_UNG", "sift1m_NHQ", "sift1m_RWalks"}
# 나머지는 모두 semi-real

# -----------------------------
# 2) 클래스별 스타일 (선/마커 통일)
# -----------------------------
STYLE_BY_CLASS = {
    "generator": {"linestyle": "-",  "marker": "D", "linewidth": 6, "markersize": 12},
    "synthetic": {"linestyle": "-",  "marker": "D", "linewidth": 6, "markersize": 12},
    "semi":      {"linestyle": "-",  "marker": "D", "linewidth": 6, "markersize": 12},
}

# -----------------------------
# 3) dataset_key → label 매핑
# -----------------------------
LABEL_MAP = {
    "sift_high":      "Generated Sift Low",
    "sift_low":       "Generated: Sift High",
    "gist_low":       "Generated Gist High",
    "gist_high":      "Generated: Gist Low",
    "sift1m_UNG":     "Synthetic: UNG",
    "sift1m_RWalks":  "Synthetic: RWalks",
    "sift1m_NHQ":     "Synthetic: NHQ",
    "sift1m_ACORN":   "Synthetic: ACORN",
    "arxiv":          "Semi-Real: arxiv",
    "LAION1M":        "Semi-Real: LAION1M",
    "tripclick":      "Semi-Real: tripclick",
    "yfcc":           "Semi-Real: YFCC",
}

# -----------------------------
# 4) 클래스별 colorblind-safe 색 팔레트
#    (Okabe–Ito & Tableau 계열, CVD-friendly)
# -----------------------------
COLOR_BY_CLASS = {
    # Generator: blue 계열 (Sky Blue, Light Blue, Blue, Darker Blue)
    "generator": [
        "#56B4E9",  # sky blue
        "#A0CBE8",  # light blue
        "#0072B2",  # blue
        "#4C72B0",  # slightly darker blue
    ],
    # Synthetic: orange/yellow 계열
    "synthetic": [
        "#E69F00",  # orange
        "#F28E2B",  # orange-ish
        "#EDC948",  # gold
        "#D55E00",  # vermillion
    ],
    # Semi-real: teal/green 계열 (CVD에서도 구분 잘 됨)
    "semi": [
        "#009E73",  # bluish green
        "#76B7B2",  # teal
        "#4DAF4A",  # green
        "#8ED1C6",  # light teal
    ],
}

# -----------------------------
# 5) trade_off를 클래스별로 그룹핑
# -----------------------------
generator_keys = [k for k in trade_off.keys() if k in GENERATOR_KEYS]
synthetic_keys = [k for k in trade_off.keys() if k in SYNTH_KEYS]
semi_keys      = [k for k in trade_off.keys() if k not in GENERATOR_KEYS | SYNTH_KEYS]

GROUPS = {
    "generator": generator_keys,
    "synthetic": synthetic_keys,
    "semi":      semi_keys,
}

# -----------------------------
# 6) 플롯
# -----------------------------
plt.figure(figsize=(12, 10))

for cls, keys in GROUPS.items():
    base_style = STYLE_BY_CLASS[cls]
    color_list = COLOR_BY_CLASS[cls]

    for local_idx, dataset_key in enumerate(keys):
        stats_dict = trade_off[dataset_key]
        label = LABEL_MAP.get(dataset_key, dataset_key)

        # 라인별 스타일 복사 (서로 영향 안 가게)
        style = base_style.copy()

        # CVD-safe 색 선택
        color = color_list[local_idx % len(color_list)]

        # UNG 강조: 테두리 + zorder 높게
        if label == "Synthetic: RWalks":
            markeredgecolor = "black"   # 여기 white로 바꿔도 좋음
            markeredgewidth = 3
            z = 10
        else:
            markeredgecolor = "none"
            markeredgewidth = 0
            z = 5

        # K_ns 정렬 (키가 "10", "20" 같은 문자열 숫자라고 가정)
        K_ns = sorted(stats_dict.keys(), key=lambda x: x)
        qps_values    = [stats_dict[k]["qps"]        for k in K_ns]
        recall_values = [stats_dict[k]["avg_recall"] for k in K_ns]

        plt.plot(
            qps_values,
            recall_values,
            marker=style["marker"],
            markersize=style["markersize"],
            markeredgecolor=markeredgecolor,
            markeredgewidth=markeredgewidth,
            linestyle=style["linestyle"],
            linewidth=style["linewidth"],
            color=color,
            label=label,
            zorder=z,
        )

plt.xscale("log")
plt.xlabel("QPS", fontsize=24, fontweight="bold")
plt.ylabel("Recall", fontsize=24, fontweight="bold")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("RWalks", fontsize=32, fontweight="bold")
plt.grid(True, which="both", linestyle=":", linewidth=0.7)

plt.legend(
    fontsize=16,
    ncol=3,
    bbox_to_anchor=(0.5, -0.15),
    loc="upper center",
)
plt.tight_layout()
plt.savefig("UNG_figure.pdf")
plt.show()
```
